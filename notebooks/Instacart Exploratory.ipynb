Few deviations from title = Name from business structure and most seem reasonable, but shows we need to de-dedup again on Title after cleaning

```python
# structure should be title = Name from business 
from_deviants = df[df["Title"].str.count("from") != 1][["Title", "Name"]]
from_deviant_index = from_deviants.index
print(from_deviants.to_string())
```

```python
def clean_title(title): 
    title = "from".join(title.split("from")[:-1])
    title = " ".join(title.split(" "))
    title = title.rstrip().lstrip()
    return title
```

```python
# clean title
df["Cleaned Title"] = df["Title"].apply(lambda x: clean_title(x))
```

```python
mask = df.duplicated(subset="Cleaned Title")
print(sum(mask))
print(sum(mask) / len(df))
df = df[~mask]
```
